```python
import library as lib
```

```python
input_path_raw = r"C:\phantichdulieu\GSE282701\GSE282701_RAW"
input_path = lib.os.path.join(lib.os.getcwd(), input_path_raw.split('\\')[-2] + '_output','adata_processed.h5ad')
adata_in_path = lib.os.path.join(lib.os.getcwd(), input_path_raw.split('\\')[-2] + '_output','adata_in.h5ad')
adata_out_path = lib.os.path.join(lib.os.getcwd(), input_path_raw.split('\\')[-2] + '_output','adata_out.h5ad')
```

```python
adata_raw = lib.sc.read_h5ad(input_path)
```

```python
adata_1 = adata_raw[adata_raw.obs['cell_type'] == 'T/NK'][:2500, :]
adata_2 = adata_raw[adata_raw.obs['cell_type'] != 'T/NK'][:2500, :]
adata = lib.ad.concat([adata_1, adata_2], axis = 0)
adata.var = adata_raw.var
adata
```

```python
adata.write_h5ad(adata_in_path)
```

```python
lib.icv.tl.infercnv(
    adata,
    reference_key="cell_type",
    reference_cat=[
    "T/NK"
    ],
    window_size=250,
)
```

```python
adata.write_h5ad(adata_out_path)
```

```python
lib.icv.tl.pca(adata)
lib.icv.pp.neighbors(adata)
lib.icv.tl.leiden(adata)
```

```python
lib.icv.pl.chromosome_heatmap(adata, groupby="cnv_leiden", dendrogram=True, save = ".png")
```

```python
lib.icv.tl.umap(adata)
lib.icv.tl.cnv_score(adata)
```
